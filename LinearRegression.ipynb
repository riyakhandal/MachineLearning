```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
%matplotlib inline
```

```python
df = pd.read_csv('brainweight.csv')
```

```python
df.head() 
```

```python
plt.scatter(df['Head Size(cm^3)'], df['Brain Weight(grams)'], color='red',marker= '*')
plt.xlabel('Head Size(cm^3)')
plt.ylabel('Brain Weight(grams)')
```

```python
reg = linear_model.LinearRegression()
```

```python
reg.fit(df[['Head Size(cm^3)']], df['Brain Weight(grams)'])
```

```python
reg.predict([[3255]])
```

```python
reg.coef_
```

```python
reg.intercept_
```
